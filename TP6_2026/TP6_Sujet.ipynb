### Mesures de dispersion des températures.

Nous souhaitons désormais mesurer la dispersion des températures, c'est-à-dire regarder comment les valeurs se répartissent autour de la moyenne. 
Les valeurs min et max ne donnent pas une bonne indication de la dispersion. 
La variance et l'écart type sont liés à la manière dont les points diffèrent de la moyenne. Plus la variance est grande, plus la dispersion est forte. 
La variance se calcule avec la formule : $$variance = \frac{\sum_i(x_i-x_{moy})^2}{n}$$ 
L'écart-type est la racine carré de la variance : $$ ect = \sqrt{variance}$$ 


- On commence par calculer la variance et l'écart type des températures de chaque mois de l'année. Il y aura donc 12 variances/écarts types. 
On fera les calculs en utilisant les opérateurs de base de numpy (np.sum(),-,/,**2), sans faire de boucle.

- Afficher les résultats obtenus.


```python
# A faire ...
tm = np.mean(t, axis=1).reshape(12,1)
var = np.sum( (t - tm)**2 , axis=1) / t.shape[1]
std = np.sqrt(var)

print(std)
```

- Utiliser ensuite les fonctions dédiées de numpy ( var et std) et vérifier que les résultats sont identiques.

```python
# A faire ...
var= np.var(t, axis=1)
std = np.std(t, axis=1)
std
```

- Tracer un graphique contenant les températures mensuelles ainsi que des barres d'erreurs indiquant la dispersion à $[-\sigma,\sigma]$ autour des moyennes où $\sigma$ désigne l'écart-type. 

```python
# A faire ...
figure, ax = plt.subplots(figsize=(16, 4))

x = np.arange(0,12)

ax.plot(x, tm.reshape(12), "b")

ax.errorbar(x, tm.reshape(12), yerr=std, capsize=5, label='Ecart temp. moyenne 1σ')

ax.legend()

ax.set(title="Analyse des temperature", xlabel='mois', ylabel='temperature')
```
