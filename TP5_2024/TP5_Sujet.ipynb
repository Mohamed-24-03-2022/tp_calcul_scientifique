### Exercice 1 : première courbe

```python
%matplotlib inline
%autosave 300
# %pylab inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from pydoc import help
```

- Tracer, sur un même graphique, le graphe des fonctions $\cos(x)$ et $\cos^2(x)$, pour 100 valeurs de $x$ comprises entre -5 et 5. 

- Ajouter les légendes et les propriétés de style de manière à ce que le résultat soit conforme à ce qui vous est donné ci-dessous. Il est obligatoire d'utiliser numpy et matplotlib.

```python
# A Faire ...
x = np.linspace(-2*np.pi, 2*np.pi, 100)
cosx = np.cos(x)
sinx = np.sin(x)


plt.figure()
plt.plot(x, cosx, "r", label="cos(x)")
plt.plot(x, cosx**2, "b", label="cos(x)^2")
plt.legend()
plt.xlim(-5,5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Mes premiers courbes")
```

- Tracer les mêmes courbes, mais cette fois-ci chacune sur un shéma sur deux zones côte à côte comme sur le shéma ci-dessous:

```python
fig, ax = plt.subplots(ncols=2, nrows=1,figsize=(12, 2))

# A Compléter ...
ax[0].plot(x, cosx, "r--", label="cos(x)")
ax[1].plot(x, cosx**2, "b", label="cos(x)^2")

ax[0].set_xlabel("x")
ax[0].set_ylabel("f(x)")
ax[0].set_title("Premiere courbe")
ax[0].legend()

ax[1].set_xlabel("x")
ax[1].set_ylabel("f(x)")
ax[1].set_title("Seconde courbe")
ax[1].legend()

```

- Reprenez les shémas ci-dessus, mais en appliquant une grille sur la figure.

```python
fig, ax = plt.subplots(ncols=2, nrows=1,figsize=(12, 2))

# A Compléter ...
ax[0].plot(x, cosx, "r--", label="cos(x)")
ax[1].plot(x, cosx**2, "b", label="cos(x)^2")

ax[0].set_xlabel("x")
ax[0].set_ylabel("f(x)")
ax[0].set_title("Premiere courbe")
ax[0].legend()
ax[0].grid(True)

ax[1].set_xlabel("x")
ax[1].set_ylabel("f(x)")
ax[1].set_title("Seconde courbe")
ax[1].legend()
ax[1].grid(True)
```
